```python
import matplotlib.pyplot as plt
import numpy as np
%pylab inline
```

```python
from dbscan_display import read_results,typenames,nametypes
```

## 5000 stars

```python
d = read_results(datatype = 'spec', case = 12, 
                 timestamp = '2018-08-22.14.01.31.93568')
d.read_base_data()
if not d.allbad:
    d.generate_average_stats(testnum=10,testsize=10,testeff=0.7,testcom=0.7,iters=100,minmem=10)

from matplotlib.colors import LogNorm
bins=20
keys = list(typenames.keys())
levels = np.linspace(0.0,1.01,bins+1)
#keys = ['spec','toph','prin30']
cmap = 'inferno'
for key in keys:
    if '{0}_statsource'.format(key) in dir(d):
        ss = getattr(d,'{0}_statsource'.format(key))
        ss = ss.data
        if any(ss['numc'] > 0.5):
            params = ss['params']
            eps = np.array([float(i.split(', ')[0]) for i in params])
            mins = np.array([int(i.split(', ')[1]) for i in params])
            mask = ss['numc'] < 0.5
            #reps = np.unique(eps)
            #rmin = mins[:len(np.unique(mins))]
            #rnum = np.reshape(ss['numc'],(len(rmin),len(reps)))
            numclevels = np.linspace(0,np.max(ss['numc'])+(np.max(ss['numc'])/bins),bins)
            ss['numc'][mask] = 0
            ss['fstd'][:,1][mask] = 0
            ss['stdeff'][:,1][mask] = 0
            ss['stdcom'][:,1][mask] = 0
            plt.figure(figsize=(20,15))
            plt.suptitle(typenames[key])
            ax = plt.subplot(221)
            ax.set_xscale('log')
            ax.set_yscale('log')
            plt.title('number of clusters found')
            #plt.tripcolor(eps,mins,ss['numc'],norm=LogNorm())
            plt.tricontourf(eps,mins,ss['numc'], bins,levels =  numclevels,cmap=cmap)
            plt.xlabel('epsilon')
            plt.ylabel('min_samples')
            plt.colorbar()
            ax = plt.subplot(222)
            ax.set_xscale('log')
            ax.set_yscale('log')
            plt.title('recovery fraction')
            #plt.tripcolor(eps,mins,ss['fstd'][:,1])#,norm=LogNorm())
            plt.tricontourf(eps,mins,ss['fstd'][:,1], bins,levels=levels,cmap=cmap)
            plt.xlabel('epsilon')
            plt.ylabel('min_samples')
            plt.colorbar()
            ax = plt.subplot(223)
            ax.set_xscale('log')
            ax.set_yscale('log')
            plt.title('efficiency')
            #plt.tripcolor(eps,mins,ss['stdeff'][:,1])#,norm=LogNorm())
            plt.tricontourf(eps,mins,ss['stdeff'][:,1],bins,levels=levels,cmap=cmap)
            plt.xlabel('epsilon')
            plt.ylabel('min_samples')
            plt.colorbar()
            ax = plt.subplot(224)
            ax.set_xscale('log')
            ax.set_yscale('log')
            plt.title('completeness')
            #plt.tripcolor(eps,mins,ss['stdcom'][:,1])#,norm=LogNorm())
            plt.tricontourf(eps,mins,ss['stdcom'][:,1],bins,levels=levels,cmap=cmap)
            plt.xlabel('epsilon')
            plt.ylabel('min_samples')
            plt.colorbar()
            
```
